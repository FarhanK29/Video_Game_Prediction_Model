```python
import matplotlib.pyplot as plt

def make_plot(df):
    plt.plot(range(len(df)),df['Players'])
    plt.ylabel("Number of Players")
    plt.title("Concurent Player Count")
    plt.xticks(rotation=0)
    plt.show()
```

```python
make_plot(watch_dogs_df)
```

```python
make_plot(watch_dogs_2_df)
```

```python
make_plot(watch_dogs_legion_df)
```

```python
def get_percent_err(df):
    df_release_player_base = df.iloc[0:1,1].values[0]
    df_end_player_base = df.iloc[-2:-1,1].values[0]

    df_percent_change = (df_end_player_base-df_release_player_base)/df_release_player_base
    return df_percent_change
```

```python
get_percent_err(watch_dogs_df)
```

```python
get_percent_err(watch_dogs_2_df)
```

```python
get_percent_err(watch_dogs_legion_df)
```
